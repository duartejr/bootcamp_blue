# Termo mais Utilizados

Neste notebook será feito uma análise dos termos mais utilizados dos produtos, e verificado quanto a qualidade e de acordo com o cluster feito anteriormente.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white')
from sklearn.feature_extraction.text import CountVectorizer
from yellowbrick.text import FreqDistVisualizer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
mercari = pd.read_csv('/content/drive/MyDrive/Datasets/dataset_com_cluster.csv')
mercari.drop('Unnamed: 0',axis=1,inplace=True)
mercari.head()
```

```python
mercari['item_condition_id'].unique()
```

## Termos de Acordo Com a Qualidade do Produto

```python
mercari.dropna(inplace=True)
count_vec = CountVectorizer(stop_words='english')
coment = count_vec.fit_transform(mercari.item_description)
features = count_vec.get_feature_names()


visu_text = FreqDistVisualizer(features=features, orient='h',size=(800, 600),
                               title='Top 50 termos mais utilizados')
visu_text.fit(coment)
visu_text.show()
```

```python
juntar_texto = ''.join(s for s in mercari.item_description)
wordcloud = WordCloud(background_color="black",width=1600, height=800).generate(juntar_texto)
```

```python
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud);
```

# Analisando outras categorias do Cluster

Neste notebook vai ser analisado outras categorias além da *women*, e coloca-las em anexo no relatório.

```python
# Importando as blibiotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importando o arquivo csv
mercari = pd.read_csv('/content/drive/MyDrive/Datasets/dataset_com_cluster.csv')
mercari.drop('Unnamed: 0',axis=1,inplace=True)
mercari.head()
```

```python
mercari['category_1'].unique()
```

## Electronics

```python
plt.figure(figsize=(10,6),dpi=150)
plt.title('Quantidade de Produtos Dentro da Categoria Electronics no Cluster 0',fontsize=14)
sns.histplot(y='category_2',data=mercari[(mercari['cluster']==0) & (mercari['category_1']=='Electronics')],
             color='royalblue')
plt.xlabel('Quantidade')
plt.ylabel('Categoria');
```
